## Data Organization

The goal is to collect correlations betwee x and y because thes.....
We will exclude free kicks from our shots and goals because we want...

```python
df['pass'] = df['eventName'] == "Pass"
df['pass'] = df['pass'].astype(int)
# df['pass']
# def get_shot(item):
#   return item == "Shot"
# df['eventName'].apply(get_shot).astype(int)
df['shot'] = df['eventName'].apply(lambda item: item=="Shot").astype(int)
df['shot'].sum()
```

```python
df['goal'] = df.apply(
    lambda row: row["eventName"] == "Shot" and "Goal" in row["tags"], 
    axis=1
    ).astype(int)
df['goal'].sum()
```

```python
df.head()
```

```python
from typing import List
def get_correlation(df,columns:List[str]):
    games_data = df.groupby(columns).agg(
        passes = ("pass", "sum"),
        shots = ("shot", "sum"),
        goals = ("goal", "sum"),
    ).reset_index()
    r,p = stats.pearsonr(games_data["passes"], games_data["shots"])
    print(f"r = {r:.4f}")
    print(f"pval = {p:.4f}")
    print(f"% of shots explained by passes: {r**2:.4%}")
```
