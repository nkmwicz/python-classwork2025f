```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"
df = pd.read_csv(url)
df.head()
```

```python
species_l = df['species'].unique()
flipper_mean = [df.loc[df['species']==species,"flipper_length_mm"].mean() for species in species_l]
plt.figure(figsize=(5,3))
plt.bar(x=species_l, height=flipper_mean, color="lightblue")
plt.show()
```

```python
bar = sns.barplot(data=df, x="species", y="flipper_length_mm")
bar.set(ylabel="Flipper Length", title="Flipper Length per Penguin Species")
```

```python
bar1 = sns.barplot(data=df, x="species", y="flipper_length_mm", hue="sex")
```

```python
plt.figure(figsize=(5,3))
box = sns.boxplot(data=df, x="species", y="body_mass_g", color="lightgray")
strip = sns.stripplot(data=df, x="species", y="body_mass_g",hue="sex")
plt.ylabel("Body Mass in g")
```

```python
hst = sns.histplot(data=df, x="body_mass_g", hue="sex", multiple="dodge")
```

```python
males_df = df.loc[df['sex'] == 'male']
females_df = df.loc[df['sex']=='female']
plt.figure(figsize=(5,3))
males = sns.histplot(data=males_df, x="body_mass_g", alpha=0.7, kde=True, label="males" )
females = sns.histplot(data=females_df, x="body_mass_g", alpha=0.7, kde=True, label="females")
m_mean = males_df['body_mass_g'].mean()
f_mean = females_df['body_mass_g'].mean()
plt.axvline(x=m_mean, label="Males mean", c='blue')
plt.axvline(x=f_mean, label="Females mean", c='orange')
plt.legend()
plt.show()
```
