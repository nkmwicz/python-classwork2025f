```python
# % pip install pandas
import pandas as pd
import random

def create_list(names_list, range_max, range_min=0):
    new_list = [
        random.choice(range(range_min,range_max)) for i in range(len(names_list))
        ]
    return new_list

# 1st way: dictionaries
names = ["Messi", "Ronaldo", "Haaland", "Mbappe", "Neymar"]
data_dict = {
    "name": names,
    "shots": create_list(names_list=names, range_max=7),
    "assists": create_list(names_list=names, range_max=4),
    "games": create_list(names_list=names, range_max=10, range_min=5)
}

df2 = pd.DataFrame(data=data_dict)
df2
```

```python
df2["goals"] = [
    random.choice(range(0,df2.at[idx,"shots"])) 
    if df2.at[idx,"shots"] != 0 else 0 
    for idx,name in enumerate(names) 
    ]
df2.head()
```
