```python
# % pip install pandas
import pandas as pd
import random

# 1st way: dictionaries
names = ["Messi", "Ronaldo", "Haaland", "Mbappe", "Neymar"]
data_dict = {
    "players": names,
    "shots": [random.choice(range(4)) for i in range(len(names))],
    "goals": [random.choice(range(7)) for i in range(len(names))],
    "assists": [random.choice(range(4)) for i in range(len(names))],
}

df1 = pd.DataFrame(data=data_dict)
df1
```

```python
# shots = [random.choice(range(7)) for i in range(len(names))]
# goals = [random.choice(range(4)) for i in range(len(names))]
# assists = [random.choice(range(4)) for i in range(len(names))]
# data = [names, shots, goals, assists]
columns = ["name", "shots", "goals", "assists"]
random.seed(22)
data = [[
    name, 
    random.normalvariate(1, 2),
    random.normalvariate(0.5, 0.7),
    random.normalvariate(0.7, 0.2),
    ] for name in names]
df2 = pd.DataFrame(data=data)
# df2 = pd.DataFrame(data=data, columns=columns)
df2
```

```python
# enumerate includes the index in the output
col_dict = {idx:col for idx, col in enumerate(columns)}
# for item in enumerate(columns):
# for idx, col in enumerate(columns): 
#     print(idx, col)
# df2.rename(columns=)
df2 = df2.rename(columns=col_dict)
df2
```

```python
# df2.rename(columns={"assists": "helpers"}, inplace=True)
df2 = df2.rename(columns={"assists": "helpers"})
df2
```

```python
df2['name']
df2['name'].values
for item in enumerate(df2['name']):
    print(item)
# print(f'then we include isn't somethin')
# df2.name
df2 = df2.rename(index={idx:name for idx, name in enumerate(df2['name'])})
# df2 = df2.reset_index()
df2
```
