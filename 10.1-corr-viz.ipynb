```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
# "https://raw.githubusercontent.com/dataprofessor/
# data/refs/heads/master/penguins_cleaned.csv"
url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"

df = pd.read_csv(url)
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
y= df['flipper_length_mm']
x= df['body_mass_g']
r,p = stats.pearsonr(x,y)
r_sq = r**2
m,b,r,p,err = stats.linregress(x,y)
#y=mx+b
# y is y value
# x is x value
# m = slope
# b = y-intercept when x=0
line = m*x + b
plt.figure(figsize=(5,3))
plt.scatter(x,y)
plt.plot(x,line,label="Line of best fit", color="red")
plt.title("Fig. 1: Body Mass to Flipper Length")
plt.xlabel("Body Mass")
plt.ylabel("Flipper Length")
text = f"""
r={r:.3f}
p={p:.3f}
r^2={r_sq:.3f}
"""
# plt.text(0.15, 0.65, text, transform=plt.gcf().transFigure)
plt.text(
    2900, 
    220, 
    text, 
    ha="left", 
    va="center",
    # bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5)
    bbox={"boxstyle": "round", "facecolor":"wheat"}
)
plt.legend()
plt.show()
```
