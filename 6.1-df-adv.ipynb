```python
import pandas as pd
import random

def create_list(names_list, range_max, range_min=0):
    new_list = [
        random.choice(range(range_min,range_max)) for i in range(len(names_list))
        ]
    return new_list

# 1st way: dictionaries
names = ["Messi", "Ronaldo", "Haaland", "Mbappe", "Neymar"]
data_dict = {
    "name": names,
    "shots": create_list(names_list=names, range_max=7),
    "assists": create_list(names_list=names, range_max=4),
    "games": create_list(names_list=names, range_max=10, range_min=5)
}

df2 = pd.DataFrame(data=data_dict)
df2
```

```python
df2["goals"] = [
    random.choice(range(0,df2.at[idx,"shots"])) 
    if df2.at[idx,"shots"] != 0 else 0 
    for idx,name in enumerate(names) 
    ]
df2.head()
```

```python
for col in df2.columns:
    if col == "games" or col == "name":
        continue
    df2[f"avg_{col}"] = df2[col] / df2["games"]

df2
```

```python
import random
# iterrows(), itertuples, items
# for item in df2.items():
    # iterates horizontally across df2
    # print(item)
# for row,cols in df2.iterrows():
    # slower of the two row-based iterators
    # print(cols['name'])
games_cols = ["name", "assists", "shots", "goals"]
data = []
for row in df2.itertuples():
    name = row.name
    assists_avg = row.avg_assists
    shots_avg = row.avg_shots
    goals_avg = row.avg_goals
    sigma = random.choice([0.5, 1.0, 1.5, 2.0, 2.5])
    for _ in range(row.games):
        shots = round(random.normalvariate(shots_avg,sigma))
        goals = round(random.normalvariate(goals_avg, sigma))
        assists = round(random.normalvariate(assists_avg, sigma))
        shots = shots if shots > 0 else 0
        goals = goals if goals > 0 else 0
        assists = assists if assists > 0 else 0
        new_row = [name, assists, shots, goals]
        data.append(new_row)

goals_df = pd.DataFrame(data=data, columns=games_cols)
goals_df.shape
        
```
